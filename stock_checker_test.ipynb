```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize a headless browser
options = webdriver.ChromeOptions()
options.add_argument("--headless")

# Provide the path to your ChromeDriver executable
driver = webdriver.Chrome(executable_path="/path/to/chromedriver", options=options)

# Navigate to the URL
url = "https://www.uniqlo.com/uk/en/product/souffl%C3%A9-yarn-jacquard-half-zipped-jumper-460948.html?dwvar_460948_size=SMA004&dwvar_460948_color=COL30"
driver.get(url)

# Find all elements with 'disabled' attribute set to 'disabled'
disabled_elements = driver.find_elements(By.XPATH, '//*[@disabled="disabled"]')

if disabled_elements:
    for element in disabled_elements:
        print(element.get_attribute("outerHTML"))

# Close the browser
driver.quit()
```
